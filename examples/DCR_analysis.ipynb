# Example on plotting the sensor population plot for LinoSPAD2

This example shows how DCR (dark count rate) can be analyzed.

Data for DCR studies should be collected in a setup that will be used in other measurements (e.g., with a source of light). However, only noise collected to characterize the DCR of the sensor directly avoiding any interference from the source.

```python
from daplis.functions import cross_talk, sensor_plot

path = r"D:\LinoSPAD2\Data\B7d\DCR_w_T"
```

First, let's take a look at the sensor population plot.

```python
# Plot sensor population plot using the first file only for a quick plot
sensor_plot.plot_sensor_population(
    path,
    daughterboard_number="B7d",
    motherboard_number="#28",
    firmware_version="2212s",
    timestamps=500,
    single_file=True,
    show_fig=True,
)
```

The plot above shows DCR, or noise, only. The peaks seen represent the hot pixels.

Now, let's go over all files and calculate the median DCR for each file.

```python
cross_talk.collect_dcr_by_file(
    path,
    daughterboard_number="B7d",
    motherboard_number="#28",
    firmware_version="2212b",
    timestamps=500,
)
```

Now that the DCR numbers were collected, we can plot them to see how DCR changed from file to file. This can be used to check for setup stability. If DCR is rising over time, it could be due to rising temperature near the sensor. 

Additionally, a histogram of the DCR over pixels is plotted to check for how different pixels contribute to DCR, for the number of hot pixels, etc.

```python
cross_talk.plot_dcr_histogram_and_stability(path)
```
